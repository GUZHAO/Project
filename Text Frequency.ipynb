```python
import urllib
from bs4 import BeautifulSoup
from bs4.element import Comment

def wordListToFreqDict(wordlist):
    wordfreq = [wordlist.count(p) for p in wordlist]
    return dict(zip(wordlist,wordfreq))

def sortFreqDict(freqdict):
    aux = [(freqdict[key], key) for key in freqdict]
    aux.sort()
    aux.reverse()
    return aux

def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True

def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)

with urllib.request.urlopen("http://news.google.com") as url:
    response = url.read()
    

dictionary = wordListToFreqDict(text_from_html(response).split())
sorteddict = sortFreqDict(dictionary)

for s in sorteddict: print(str(s))
```

```python
from bs4 import BeautifulSoup
import requests
#import urllib
import os
import pprint
import json

#with urllib.request.urlopen('http://www.upneat.rocks/recipe/sources/death-co') as url:
#    r = url.read()
r = requests.get('http://www.upneat.rocks/recipe/sources/death-co').text
soup = BeautifulSoup(r,"html.parser")
recipes = soup.find_all("div",class_="col-xs-12 col-md-6 col-lg-3")
#print(soup.prettify())
#print(recipes)
#recipes[4].a["href"]

lobbying={}
for element in recipes:
    lobbying[element.a.get_text()]={}
    
prefix = "http://www.upneat.rocks"
for element in recipes:
    lobbying[element.a.get_text()]["link"] = prefix + element.a["href"]
print(lobbying)
    
for item in lobbying.keys():
    print (item + ": " + "\n\t" + "link: " + lobbying[item]["link"] + "\n\n") 

os.chdir("C:/Users/gz056")

with open("lobbying.json", "w") as writeJSON:
    json.dump(lobbying, writeJSON)

with open("C:/Users/gz056/lobbying.json") as data_file:    
    data = json.load(data_file)
pprint.pprint(data)
```
